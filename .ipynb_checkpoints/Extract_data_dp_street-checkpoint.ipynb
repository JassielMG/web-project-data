```python
import numpy as np
import requests
import pandas as pd
from bs4 import BeautifulSoup

class Spider:
      
    def __init__(self, url_pattern, pages_to_scrape=10, sleep_interval=6, content_parser=None):
        self.url_pattern = url_pattern
        self.pages_to_scrape = pages_to_scrape
        self.sleep_interval = sleep_interval
        self.content_parser = content_parser
       
    
   
    def scrape_url(self, url,j):
        header = {
                'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
                'referrer': 'https://google.com',
                'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
                'Accept-Encoding': 'gzip, deflate, br',
                'Accept-Language': 'en-US,en;q=0.9',
                'Pragma': 'no-cache',}
        try:
            response = requests.get(url,headers=header)
            if response.status_code == 200:
                print('connection established code 200')
                result = self.content_parser(response.content,j)
                self.output_results(result)
            else:
                raise ServerError
        except:
            print('Server error, Failed Connection')
            
        
    def output_results(self, r):
        print(r)
    

    def kickstart(self):
        for j,i in enumerate(range(0, self.pages_to_scrape*24,24)):
            t = np.random.choice(range(0,self.sleep_interval))
            time.sleep(t)
            self.scrape_url(self.url_pattern % i,j)


URL_PATTERN = 'https://www.dpstreet.mx/SearchDisplay?categoryId=16051&storeId=10152&catalogId=13051&langId=-5&sType=SimpleSearch&resultCatEntryType=2&showResultsPage=true&searchSource=Q&pageView=&beginIndex=0&pageSize=24&searchTerm=tenis#facet:&productBeginIndex:%s&orderBy:&pageView:grid&minPrice:&maxPrice:&pageSize:&'
PAGES_TO_SCRAPE = 10 # how many webpages to scrapge

data_extract=[]
def books_parser(content,j):
    
    html = content
    soup = BeautifulSoup(html,'lxml')
    
    tenis = soup.select("div[class='product_name'] a")
    tenis = [teni.text[6:-7] for teni in tenis]
    
    precios = soup.select("div[class='product_price']>span:nth-child(1)")
    precios = [p.text.strip().replace('Precio','').replace('Oferta','').strip(':').strip('\r\n').strip('\t').strip() for p in precios]
    
    images = soup.select("div[class='image'] a img")
    images = [image['src'] for image in images]
    
    links = soup.select("div[class='product_name'] a")
    links = [link['href'] for link in links]
    
    modelos=[]
    codigos=[]
    for url in links:
        html = requests.get(url).content
        subsoup = BeautifulSoup(html,'lxml')
        modelo = subsoup.select("div[class='product_text'] dl>dd:nth-child(2)")[0].text
        codigo = subsoup.select("div[class='product_text'] dl>dd:nth-child(4)")[0].text
        modelos.append(modelo)
        codigos.append(codigo)
    
    df = pd.DataFrame({
        'Sneaker':tenis,
        'Modelo':modelos,
        'Precio':precios,
        'Codigo':codigos,
        'Image':images,
        'Url':links})  
    
    data_extract.append(df)
    
    return f'Los datos de la pagina {j+1} del catalogo han sido extraidos.. '


# Instantiate the IronhackSpider class
my_spider = Spider(URL_PATTERN, PAGES_TO_SCRAPE, content_parser=books_parser)

# Start scraping jobs
my_spider.kickstart()
```

```python
total_data = pd.concat(data_extract,axis=0).reset_index(drop=True)
total_data.head(6)
```

```python
total_data.to_csv('Sneakers.csv')
```
